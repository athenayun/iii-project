```python
# innisfree

import requests as r
from bs4 import BeautifulSoup
import shutil


pro_url = "http://www.innisfreeworld.com/product/productView.do?prdSeq=10832&eventSeq=0&catCd01=UB"
res = r.get(pro_url)

res.encoding = 'utf-8' # 調整Response obect的encoding，不然中文是亂碼
soup = BeautifulSoup(res.text, 'html5lib')

title = "http://www.innisfreeworld.com"

res_links = []
host = "http://www.innisfreeworld.com"

res_links.extend([host + link['src'] for link in soup.select("li > a[href^='#'] > img")])


names = [link['alt'] for link in soup.select("li > a[href^='#'] > img")]



for i in range(len(names)):
    res2 = r.get(res_links[i], stream=True)
    with open('./color_temp/%s.jpg'%names[i], 'wb') as f:
            shutil.copyfileobj(res2.raw, f)
```

```python
# RMK

import requests as r
from bs4 import BeautifulSoup

pro_url = "http://www.ilovermk.com.tw/product-detail.asp?ids=396"
#pro_url = "http://www.ilovermk.com.tw/product-detail.asp?ids=395"

res = r.get(pro_url)

res.encoding = 'utf-8' # 調整Response obect的encoding，不然中文是亂碼
soup = BeautifulSoup(res.text, 'html5lib')


res_links = [link['src'] for link in soup.select("li > a > img[src^='http']")]
names = [link['alt'] for link in soup.select("li > a > img[src^='http']")]


temp = []
for i in names:
    name = i.split('：')[1]
    temp.append(name)

for i in range(len(temp)):
    res2 = r.get(res_links[i], stream=True)
    with open('./color_temp/%s.jpg'%temp[i], 'wb') as f:
            shutil.copyfileobj(res2.raw, f)
```

```python
# CANMAKE

import requests as r
from bs4 import BeautifulSoup

pro_url = "http://www.canmake.com/en/elip/item_lip025.html"
#pro_url = "http://www.canmake.com/en/elip/item_lip021.html"
res = r.get(pro_url)

res.encoding = 'utf-8' # 調整Response obect的encoding，不然中文是亂碼
soup = BeautifulSoup(res.text, 'html5lib')

host = "http://www.canmake.com"
temp = [link['src'] for link in soup.select("div > img[alt$='image']")]
res_links = []
for i in temp:
    dada = host + i.split('..')[2]
    res_links.append(dada)


b = '%s'%soup.select("div > .color_name")


c=b.split(",")

names = []
for i in c :
    d = i.split('<div class="color_name">')[1].split('</div>')[0]
    names.append(d)
    
for i in range(len(names)):
    res2 = r.get(res_links[i], stream=True)
    with open('./color_temp/%s.jpg'%names[i], 'wb') as f:
            shutil.copyfileobj(res2.raw, f)
```

```python
# Givenchy

import requests as r
from bs4 import BeautifulSoup

pro_url = "https://www.givenchybeauty.com/en/make-up/lipstick/range_1_1"
#pro_url = "https://www.givenchybeauty.com/en/make-up/lipstick/range_1"
res = r.get(pro_url)

res.encoding = 'utf-8' # 調整Response obect的encoding，不然中文是亂碼
soup = BeautifulSoup(res.text, 'html5lib')

res_links = [link['src'] for link in soup.select("div > p.shade > img[src^='https']")]

soup.select_one("div > p.shade").text
names = []

for i in soup.select("div > p.shade"):
    a = i.text
    names.append(a.split('\n')[1])
    
for i in range(len(names)):
    res2 = r.get(res_links[i], stream=True)
    with open('./color_temp/%s.jpg'%names[i], 'wb') as f:
            shutil.copyfileobj(res2.raw, f)
```
