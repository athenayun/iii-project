```python
#find the keyword in each product
import codecs
import json
import jieba.analyse

with codecs.open('../eyebrow.txt','r') as s:
    jieba.load_userdict(s)


with open('../MAYBELLINE 媚比琳_宛若真眉雙效眉粉筆_one_color.json') as f:
    test=json.load(f)

with codecs.open('./stopwords.txt','r','utf8') as f:
    
    stopwords=f.read().split('\n')
    
dic={}
i=0
while True:
    try:
        test1=test['comments'][i]['content']
    except IndexError :
        break
    words=jieba.lcut(test1)
    for word in words  :
        word=(word.replace('～','')
                  .replace('~','')
                  .replace('↓','')
                  .replace('!','。')
                  .replace('(','')
                  .replace(')','')
                  .replace('\xa0',',')
                  .replace('...',',')
                  .replace(' ）','')
                  .replace('?','')
             )     
      
        if word in stopwords:
            continue
        if word in dic:
            dic[word]+=1
        else :
            dic[word]=0
    i+=1
sorted(dic.items(),key=lambda x : x[1],reverse=True)
    
    
    
```
