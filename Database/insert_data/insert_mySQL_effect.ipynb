```python
class effect(peewee.Model):
    pro_id = peewee.IntegerField()
    pro_class = peewee.IntegerField()
    Coloring = peewee.FloatField()
    Moisture = peewee.FloatField()
    Durability = peewee.FloatField()
    Pushing_evenness = peewee.FloatField()
    Waterproof = peewee.FloatField()
    Distinctness = peewee.FloatField()
    Naturalness = peewee.FloatField()
    Transparency = peewee.FloatField()
    Saturation = peewee.FloatField()
    Concealer = peewee.FloatField()
    Sticking = peewee.FloatField()
    Refreshing = peewee.FloatField()
    
    class Meta:
        database = db
```

```python
def unified_effect_field(j_file):
    
    if 'Coloring' not in j_file.keys():
        j_file['Coloring'] = 0
    
    if 'Moisture' not in j_file.keys():
        j_file['Moisture'] = 0

    if 'Durability' not in j_file.keys():
        j_file['Durability'] = 0
    
    if 'Pushing_evenness' not in j_file.keys():
        j_file['Pushing_evenness'] = 0
        
    if 'Waterproof' not in j_file.keys():
        j_file['Waterproof'] = 0
        
    if 'Distinctness' not in j_file.keys():
        j_file['Distinctness'] = 0
        
    if 'Naturalness' not in j_file.keys():
        j_file['Naturalness'] = 0
        
    if 'Transparency' not in j_file.keys():
        j_file['Transparency'] = 0
        
    if 'Saturation' not in j_file.keys():
        j_file['Saturation'] = 0
        
    if 'Concealer' not in j_file.keys():
        j_file['Concealer'] = 0
        
    if 'Sticking' not in j_file.keys():
        j_file['Sticking'] = 0
        
    if 'Refreshing' not in j_file.keys():
        j_file['Refreshing'] = 0
        
    #changing class name    
    
    if j_file['pro_class'] == '眉筆' :
        j_file['pro_class'] = '1'
    elif j_file['pro_class'] == '眼線筆':
        j_file['pro_class'] = '2'
    elif j_file['pro_class'] == '粉底液':
        j_file['pro_class'] = '3'
    elif j_file['pro_class'] == '唇膏':
        j_file['pro_class'] = '4'
    else:
        print("out of classification")
    
    
        
    
        
    return j_file
```

```python
# changing score dictionary items

import glob
import json
import codecs
from pprint import pprint

files_base = glob.glob('D:\\project\\urcosme\\Score\\*.json')



value_lists = []


for i in files_base :
    
    with codecs.open(i,'r','utf-8')as file:
        
        j_rawfile = json.load(file)
        
        j_items = j_rawfile.items()
        
        for j in j_items:
            
            j_value = j[1]
            value_lists.append(j_value)
            
        for i in value_lists:
            unified_effect_field(i)
        
        
            
        #print(value_lists)
```
