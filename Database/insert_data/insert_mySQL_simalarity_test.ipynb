```python
#  connection on localhost

import peewee
import pymysql

db = peewee.MySQLDatabase('myproject',
                          user='root',
                          host='localhost',
                          password='athena1225')
```

```python

class similarity(peewee.Model):
    product_id_1 = peewee.IntegerField()
    product_id_2 = peewee.IntegerField()
    similar = peewee.FloatField()
    appearence = peewee.FloatField()
    
   
    
    class Meta:
        database = db
```

```python

def insert_data(df):
    dicts=({'product_id_1': df['product1'][i],
            'product_id_2': df['product2'][i],
            'similar': float(df['similar'][i]),
            'appearence': float(df['appearence'][i]),
            
           } for i in range(df['product1'].count()))
    with db.atomic():
        similarity.insert_many(dicts).execute()
```
